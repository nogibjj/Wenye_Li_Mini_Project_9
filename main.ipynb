```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset_url = "https://raw.githubusercontent.com/fivethirtyeight/data/refs/heads/master/drug-use-by-age/drug-use-by-age.csv"
```

```python
def load_dataset():
  return pd.read_csv(dataset_url)

data = load_dataset()
assert data is not None
assert data.shape == (17, 28)
data.head()
```

```python
def describe_dataset():
    return pd.read_csv(dataset_url).describe()

description = describe_dataset()
description
```

```python
mean = description.mean()
assert mean is not None
print("The mean of the dataset is: ")
mean
```

```python
std = description.std()
assert std is not None
print("The std of the dataset is: ")
std
```

```python
def describe_median():
  dataset = load_dataset().replace("-", 0)
  median_values = dataset.iloc[:, 1:].median()
  return median_values

median_values = describe_median()
assert median_values is not None
print("The median of the dataset is: ")
median_values
```

```python
plt.figure(figsize=(8, 6))

# Plotting the histogram
plt.hist(data['alcohol_use'], bins=10, edgecolor='black')
plt.xlabel('Alcohol Use (%)')
plt.ylabel('Frequency')
plt.title('Distribution of Alcohol Use')

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
plt.close()
```

```python
data_selected = data[['age', 'marijuana_use']]

plt.figure(figsize=(10, 6))
plt.plot(data_selected['age'], data_selected['marijuana_use'], label='Marijuana Use (%)', marker='o', color='g')

plt.xlabel('Age')
plt.ylabel('Marijuana Use (%)')
plt.title('Trend of Marijuana Use by Age')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
plt.close()
```

```python
data_selected = data[['age', 'cocaine_use']]

plt.figure(figsize=(10, 6))
plt.bar(data_selected['age'], data_selected['cocaine_use'], color='orange')

plt.xlabel('Age')
plt.ylabel('Cocaine Use (%)')
plt.title('Cocaine Use by Age')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.close()
```
